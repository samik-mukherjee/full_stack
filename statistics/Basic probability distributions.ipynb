## Uniform Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-10,10,10000)
plt.hist(values, 50)
plt.show()
```

## Normal Distribution 

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-3,5,0.001)
plt.plot(x, norm.pdf(x), 'r--')
plt.plot(x, norm.pdf(x, 1.0,1.5), 'b-') # x axes and norm.pdf(x) for y axes
```

## Moments : Mean, Variance, Skew, Kurtosis

```python
import numpy as np 
import matplotlib.pyplot

# mean around 0, standard deviation 0.5, 1000 data points
values = np.random.normal(0,0.5,1000) 

plt.hist(values, 50)
plt.show()
```

The first moment is the mean, this data should average about 0.

```python
np.mean(values) # finding the mean of values
```

The second moment is vatiance which is just the squre of standerd deviation

```python
np.var(values) # squre of the standerd deviation or variance
```

The third moment is skew- this data is centered around 0, so skew should be close to 0

```python
import scipy.stats as sp
sp.skew(values) 
```

The fourth moment is kurtosis which describes the shape of the tail, for nornal distribution it is close to 0

```python
sp.kurtosis(values) # shape of the tail 
```
